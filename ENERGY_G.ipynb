```python
import os
os.environ["JAX_ENABLE_X64"] = "True"
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import sys
import pandas as pd
from matplotlib import cm
import matplotlib as mp
import matplotlib.colors as mcolors
import scipy.special as sc
import scipy.optimize as so
from matplotlib.colors import LinearSegmentedColormap
```

```python
L=10
W=1
GI=50
GF=70
DG=5
NS=64
NR=100
n_mean=10
start=0
method=4
norm = mcolors.Normalize(vmin=np.abs(GI), vmax=np.abs(GF))
name_figs="L"+str(L)+"W"+str(W)+"NS"+str(NS)+"NR"+str(NR)+"GI"+str(GI)+"GF"+str(GF)+"D.png"
```

```python
G=[i for i in range(GI,GF+DG,DG)]
name="M"+str(method)+"L"+str(L)+"W"+str(W)+"NS"+str(NS)+"NR"+str(NR)+"G"+str(GI)+"GF"+str(GF)+"GE.txt"
file=pd.read_csv(name,sep="\s+",dtype="a")
file=file.astype(float)  
```

```python
G=np.array(G)
plt.figure()
plt.title(r'$\;E$ '+"\n"+r" $L="+str(L)+"$ "+r" $\; N_{average}="+str(n_mean)+"$",fontsize="15")
plt.errorbar(G*0.01,file["E"],yerr=file["dE"],marker="o")
plt.xlabel('$E$',fontsize="15")
plt.ylabel(r'$S$',fontsize="15")
plt.show()
```
