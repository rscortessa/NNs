```python
import os
os.environ["JAX_ENABLE_X64"] = "True"
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import sys
import pandas as pd
from matplotlib import cm
import matplotlib as mp
import matplotlib.colors as mcolors
import scipy.special as sc
import scipy.optimize as so
from matplotlib.colors import LinearSegmentedColormap
import Methods.class_WF as class_WF
```

```python
L=48
W=1
GI=90
NI=100
NR=10
NF=1000
NS=1024
NNs=1
n_mean=90
start=0
method=3
model="QIM"
name_method="RBM"
norm = mcolors.Normalize(vmin=np.abs(NI), vmax=np.abs(NF))
name_figs="L"+str(L)+"W"+str(W)+"G"+str(GI)+"NN"+str(NNs)+"NS"+str(NS)+"NI"+str(NS)+"NF"+str(NF)+"S.png"
```

```python
G=[1/i for i in range(NI,NF+NI,NI)]
name="M"+str(method)+"L"+str(L)+"W"+str(W)+"NS"+str(NS)+"NB"+str(NI)+"G"+str(GI)+"NR"+str(NR)+"NSS.txt"
file=pd.read_csv(name,sep="\s+",dtype="a")
file=file.astype(float)  
```

```python
name_teo="DATAM"+str(5)+"L"+str(L)+"W"+str(W)+"NS"+str(1000)+"MPSG"+str(GI*10)+".txt"
file_teo=pd.read_csv(name_teo,sep="\s+",dtype="a")
file_teo=file_teo.astype(float)
A=np.array(file_teo)
lenght=len(A)
A[:int(lenght/2),:]=(-1)*A[:int(lenght/2),:]
eps=10**(-16)
S_teo=class_WF.S_PCA(A,eps,False)
```

```python
plt.figure()
plt.title(r'$\% \;S_{PCA}$ '+"\n"+r" $L="+str(L)+"$ "+r" $\; N_{average}="+str(n_mean)+"$"+r"$\;h/J="+str(GI*0.01)+"$",fontsize="15")
plt.errorbar(G,np.abs(file["S"]-S_teo)/S_teo,yerr=file["dS"]/np.sqrt(n_mean),marker="o")
plt.xlabel('$1/N_{steps}$',fontsize="15")
plt.ylabel(r'$\% S_{PCA}$',fontsize="15")
plt.xscale("log")
plt.tick_params(axis='both', which='major', labelsize=20, width=2.5, length=10)
plt.savefig(name_figs)
plt.show()
```

```python
plt.figure()
plt.title(r'$\;S_{PCA}$ '+"\n"+r" $L="+str(L)+"$ "+r" $\; N_{average}="+str(n_mean)+"$"+r"$\;h/J="+str(GI*0.01)+"$",fontsize="15")
plt.errorbar(G,(file["S"]),yerr=file["dS"]/np.sqrt(n_mean),marker="o")
plt.xlabel('$1/N_{steps}$',fontsize="15")
plt.ylabel(r'$S$',fontsize="15")
plt.xscale("log")
plt.tick_params(axis='both', which='major', labelsize=20, width=2.5, length=10)
plt.savefig(name_figs)
plt.show()
```
